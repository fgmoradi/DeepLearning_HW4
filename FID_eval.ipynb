```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DCGAN = np.load('Results/FID_score/DC_FID.npy')
WGAN = np.load('Results/FID_score/WGAN_FID.npy')
ACGAN = np.load('Results/FID_score/ACGAN_FID.npy')
```

```python
plt.figure(figsize=(10,5))
plt.title("FID Scores for DCGAN vs WGAN vs WGANGP vs ACGAN")
plt.plot(DCGAN,label="DCGAN")
plt.plot(WGAN,label="WGAN")
plt.plot(ACGAN,label="ACGAN")
plt.xlabel("Epochs")
plt.ylabel("FID")
plt.legend()
plt.show()
plt.savefig('Results/FID_summary.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```

```python
import numpy as np
import pandas as pd
data = {
    "Model": ["DCGAN", "WGAN", "ACGAN"],
    "Min FID": [
        np.min(DCGAN),
        np.min(WGAN),
        np.min(ACGAN),
    ],
    "Mean FID": [
        np.mean(DCGAN),
        np.mean(WGAN),
        np.mean(ACGAN),
    ],
    "FID @ epoch 50": [
        DCGAN[49],
        WGAN[49],
        ACGAN[49],
    ],
}

df = pd.DataFrame(data)

# Show the table (rounded) in the notebook
print(df.round(2))

# If you want Markdown for your report/notebook:
print("\nMarkdown table:\n")
print(df.round(2).to_markdown(index=False))
```
